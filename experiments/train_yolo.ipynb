Notebook for training YOLOv8 on the KozAlma dataset.

```python
from ultralytics import YOLO

# Load base model
model = YOLO('yolov8n.pt')

# Train
results = model.train(
    data='../data/data.yaml',
    epochs=100,
    imgsz=640,
    batch=16,
    name='koz_alma_train',
    patience=20,
    save=True,
    plots=True,
)
```
